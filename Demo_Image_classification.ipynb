```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = load_model('/content/drive/MyDrive/Image classification Project/Skinlesion_model.h5')
```

```python
image_path = '/content/drive/MyDrive/Image classification Project/To predict_nv.png'
```

```python
img = image.load_img(image_path, target_size=(224, 224))
```

```python
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0
```

```python
predictions = model.predict(img_array)
```

```python
class_names = ['nv', 'bkl', 'mel', 'bcc']
predicted_class = class_names[np.argmax(predictions)]
confidence = np.max(predictions) * 100
```

```python
print(f"Classe predetta: {predicted_class}")
print(f"Confidenza: {confidence:.2f}%")

plt.figure(figsize=(8, 8))
img = image.load_img(image_path)
plt.imshow(img)
plt.axis('off')
plt.title(f"Predizione: {predicted_class} ({confidence:.2f}%)")
plt.show()
```
